The JRC-Arquis provides the files in XML format, therefore we have to load and process the XML files

```python
import xml.etree.ElementTree as ET
import glob
import numpy as np
import pandas as pd
import os
```

Example for the first file: It has some tree structure, so we have a root node and children 

```python
path = "/content/en"
example_folder = os.listdir(path)[0]
example_file = os.listdir(path+"/"+example_folder)[0]
print("Example file: {}".format(example_file))

#Extract some sample data from the file
tree = ET.parse(path+"/"+example_folder+"/"+example_file)
root = tree.getroot()
print("Title: {}".format(root[0][0][0][0].text))
body = ''.join(root[1].itertext())
print("Content: {}...".format(body[:100]))
```

Now define a function to extract the text from all the files and load them into pandas DataFrames

```python
def read_convert_whole(path, lang):
  c=[]
  for folder in os.listdir(path):
    for filename in os.listdir(path+"/"+folder):
        tree = ET.parse(path+"/"+folder+"/"+filename)
        root = tree.getroot()
        number=root.attrib["n"]
        title=root[0][0][0][0].text
        # title_long=root[0][0][0][1].text
        body=''.join(root[1].itertext())
        c.append([number, body]) #,title,body])
  df = pd.DataFrame (c,columns=['number', "body"]) #'lang','title','body']) 
  return df
```

```python

de=read_convert_whole("/content/de","de")
en=read_convert_whole("/content/en","en")
nl=read_convert_whole("/content/nl","nl")
pl=read_convert_whole("/content/pl","pl")
hu=read_convert_whole("/content/hu","hu")
cs=read_convert_whole("/content/cs","cs")
```

```python
from functools import reduce
data_frames = [de, en, nl, pl, hu, cs]
df_merged = reduce(lambda left,right: 
                    pd.merge(left,right, on=['number']), data_frames)

df_merged.columns = ['number',
                    'body_de',
                    'body_en', 
                    'body_nl',
                    'body_pl',
                    'body_hu',
                    'body_cs']
```
